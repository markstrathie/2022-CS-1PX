```python
#Q1.
def read_data(filename):
    #setting up dictionaries that will be returned
    dataDictionary = {}
    dictionaryCountry = {}
    dictionaryYear = {}
    #try block - exception catching
        #open file and skip top 2 lines which are just detailed column labels
    f = open(filename, "r")
    f.readline()
    f.readline()
        #read data line-by-line, stripping and splitting on commas
    for line in f:
        split = line.strip().split(',')
            # if-else - error checking for missing columns of data
        if len(split) == 8:
                #labelling data and slicing off punctuation and unused data
            country = str(split[0][1:-1])
            year = int(split[1][1:-1])
            values = [float(split[2][1:5]),float(split[3][1:5]),float(split[4][1:5]),float(split[5][1:5]),float(split[6][1:5]),float(split[7][1:5])]
                #building data structure
            keys = ['ProbDeathBoth', 'ProbDeathMale', 'ProbDeathFemale', 'SuicideRateBoth', 'SuicideRateMale', 'SuicideRateFemale']
            i = 0
            for key in keys:
                dictionaryYear[key] = values[i]
                i += 1
            dictionaryCountry[year] = dictionaryYear
            dataDictionary[country] = dictionaryCountry
            # if-else - error checking for missing columns of data
        else:
            print('Missing data on line ' + line)
            return dataDictionary
    return dataDictionary
    #except block - exception catching

read_data('data.csv')
```

```python
#Q1.
#filter by country
def filter_by_characteristic(dataStruct, countryString):
    #setting up empty dictionary for filtered data
    newDictionary = {}
    #error-checking, making sure input is a string. If False user will receive message
    assert type(countryString) == str, 'Please input country as a string'
    #for loop, checking dataset for matching country and if yes adding to new dictionary
    for country in dataStruct:
        if country == countryString:
            newDictionary[country] = dataStruct[country]
    #error-checking, inform user if no country is found, therefore new dictionary is empty
    if newDictionary == {}:
        print('Unable to locate country')
    return newDictionary

#Q2.
#average suicide rate worldwide, per year
def aggregate_by_characteristic(dataStruct, SuicideRate_Both_orMale_orFemale_String):
    #setting up new list for tuples containing year and value to be added
    newList = []
    #setting up while loop to run from year 2019 down to 1999
    year = 2019
    while year > 1999:
        #setting up variables for sum and average to be calculated
        sum = 0
        average = 0
        #for loop to find values for current year and calculate average
        for country in dataStruct:
            sum += dataStruct[country][year][SuicideRate_Both_orMale_orFemale_String]
        average = sum / len(dataStruct)
        #error-checking, making sure average is a valid result, informing user which year error occurs if error
        assert average > 5 and average < 40, 'Error located with data in year ' + year
        #creating tuple, adding to new list, decrementing year
        tup = (year, average)
        newList.append(tup)
        year -= 1
    #error-checking, informing user if returned list is not full 20 years of data
    if len(newList) != 20:
        print('Result does not contain years 2000-2019 please check data')
        return newList
    return newList
```
