Task B: Reading in your data ***(total 15 marks)***
1. write a function called `read_data(filename)` that takes your .csv `filename` as an argument, and `returns` an instance of the data structure you have devised.  You should use at least one example of defensive error-checking and one example of exception-catching.

```python
#Q1.
def read_data(filename):
    #setting up dictionaries that will be returned
    dataDictionary = {}
    dictionaryCountry = {}
    dictionaryYear = {}
    #try block - exception catching
    try:
        #open file and skip top 2 lines which are just detailed column labels
        f = open(filename, "r")
        f.readline()
        f.readline()
        #read data line-by-line, stripping and splitting on commas
        for line in f:
            split = line.strip().split(',')
            # if-else - error checking for missing columns of data
            if len(split) == 8:
                #labelling data and slicing off punctuation and unused data
                country = str(split[0][1:-1])
                year = int(split[1][1:-1])
                ProbDeathBoth = float(split[2][1:5])
                ProbDeathMale = float(split[3][1:5])
                ProbDeathFemale = float(split[4][1:5])
                SuicideRateBoth = float(split[5][1:5])
                SuicideRateMale = float(split[6][1:5])
                SuicideRateFemale = float(split[7][1:5])
                #building data structure
                dictionaryYear = {'ProbDeathBoth': ProbDeathBoth, 'ProbDeathMale': ProbDeathMale, 'ProbDeathFemale': ProbDeathFemale, 'SuicideRateBoth': SuicideRateBoth, 'SuicideRateMale': SuicideRateMale, 'SuicideRateFemale': SuicideRateFemale}
                dictionaryCountry[year] = dictionaryYear.copy()
                dataDictionary[country] = dictionaryCountry.copy()
            # if-else - error checking for missing columns of data
            else:
                print('Missing data on line ' + line)
                return dataDictionary
        return dataDictionary
    #except block - exception catching
    except:
        print('Unable to parse line ' + line)
        return dataDictionary

dataStruct = read_data('data.csv')
```

```python
#Q1.
#filter by country
def filter_by_characteristic(dataStruct, countryString):
    #setting up empty dictionary for filtered data
    newDictionary = {}
    #error-checking, making sure input is a string. If False user will receive message
    assert type(countryString) == str, 'Please input country as a string'
    #for loop, checking dataset for matching country and if yes adding to new dictionary
    for country in dataStruct:
        if country == countryString:
            newDictionary[country] = dataStruct[country]
    #error-checking, inform user if no country is found, therefore new dictionary is empty
    if newDictionary == {}:
        print('Unable to locate country')
    return newDictionary

UK = filter_by_characteristic(dataStruct, 'United Kingdom of Great Britain and Northern Ireland')

#Q2.
#average suicide rate worldwide, per year
def aggregate_by_characteristic(dataStruct, SuicideRate_Both_orMale_orFemale_String):
    #setting up new list for tuples containing year and value to be added
    newList = []
    #setting up while loop to run from year 2019 down to 1999
    year = 2019
    while year > 1999:
        #setting up variables for sum and average to be calculated
        sum = 0
        average = 0
        #for loop to find values for current year and calculate average
        for country in dataStruct:
            sum += dataStruct[country][year][SuicideRate_Both_orMale_orFemale_String]
        average = sum / len(dataStruct)
        #error-checking, making sure average is a valid result, informing user which year error occurs if error
        assert average > 0 and average < 100, 'Error located with data in year ' + str(year)
        #creating tuple, adding to new list, decrementing year
        tup = (year, average)
        newList.append(tup)
        year -= 1
    #error-checking, informing user if returned list is not full 20 years of data
    if len(newList) != 20:
        print('Result does not contain years 2000-2019 please check data')
        return newList
    return newList

WorldwideRate = aggregate_by_characteristic(dataStruct, 'SuicideRateBoth')
```

```python
#Q1.
import matplotlib.pyplot as plt

def scatter_data(dataStruct):
    #for loop to run through dictionary twice, first for Male data and second for Female data
    for i in range(2):
        #if-else to index dictionary correctly and label data for graph
        if i == 0:
            data = 'SuicideRateMale'
            labelSex = 'Male'
        else:
            data = 'SuicideRateFemale'
            labelSex = 'Female'
        #setting up lists for x and y values for graph
        xVals = []
        yVals = []
        yearInt = 2019
        #while loop to loop through years 2019 - 2000 and add data to lists
        while yearInt > 1999:
            for country in dataStruct:
                xVals.append(dataStruct[country][yearInt]['ProbDeathBoth'])
            for country in dataStruct:
                yVals.append(dataStruct[country][yearInt][data])
            yearInt -= 1
        plt.scatter(xVals, yVals, s=1, label=labelSex)
    
    #title, axis labels, legend for graph
    plt.title("Worldwide Suicide Rate 2000 - 2019")
    plt.xlabel("Probability of dying between 30 and 70 from cardiovascular disease,\ncancer, diabetes, or chronic respiratory (%) for Both Sexes")
    plt.ylabel("Suicide mortality rate\n(per 100 000 population)")
    plt.legend(loc="upper left")
    plt.show()
    
scatter_data(dataStruct)
    
#Q2.
#histogram to show distribution of suicide rate over all countries in selected year
def hist_data(dataStruct, yearInt, dataString):
    #setting empty list for data needed for graph
    vals = []
    #for loop to traverse dictionary and select chosen data for chosen year for every country
    for country in dataStruct:
        vals.append(dataStruct[country][yearInt][dataString])
    plt.hist(vals, color='green')
    
    #title, axis labels, show graph
    plt.title("Worldwide Suicide Rate " + str(yearInt))
    plt.xlabel("Suicide mortality rate\n(per 100 000 population) for Both Sexes")
    plt.ylabel("Number of countries")
    plt.show()

hist_data(dataStruct, 2019, 'SuicideRateBoth')

#Q3.
#pie chart to show percentage of male vs female suicide rate in UK in 2019.
def pie_data(dataStruct, yearInt, countryString):
    #filtering by country to UK
    country = filter_by_characteristic(dataStruct, countryString)
    #locating Male and Female rates for selected year
    maleRate = country[countryString][yearInt]['SuicideRateMale']
    femaleRate = country[countryString][yearInt]['SuicideRateFemale']
    #converting to percentage and making list of both values
    total = maleRate + femaleRate
    data = []
    data.append((maleRate / total) * 100)
    data.append((femaleRate / total) * 100)
    #creating chart
    myLabels = ['Male', 'Female']
    plt.pie(data, labels=myLabels, startangle=90)
    plt.title("Percentage of Male against Female Suicide\nin " + countryString + "\nduring " + str(yearInt))
    plt.show()

pie_data(dataStruct, 2019, 'United Kingdom of Great Britain and Northern Ireland')    
```
